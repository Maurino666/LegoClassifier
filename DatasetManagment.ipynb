```python
import os
import csv
from pathlib import Path
```

Prendo le classi disponibili dalla cartella da D1

```python
# Percorso di D1
percorso_d1 = "Dataset/D1"  # Percorso relativo

def estrai_classi_cartelle(percorso):
    return [item.name for item in Path(percorso).iterdir() if item.is_dir()]

classi_d1 = estrai_classi_cartelle(percorso_d1)
print(len(classi_d1))
print(classi_d1)
```

Prendo le classi da D2

Organizzo le classi in cartelle separate.

```python
import shutil

# Percorso di D2
percorso = "Dataset/D2tofix/dataset"

def organizza_file_per_classe(percorso):
    percorso_base = Path(percorso)

    for file in percorso_base.iterdir():
        # Controlla se  un file
        if file.is_file():
            # Estrarre il codice della classe (prima dello spazio)
            codice_classe = file.name.split(' ')[0]
            # Percorso della cartella della classe
            cartella_classe = percorso_base / codice_classe

            # Crea la cartella se non esiste
            cartella_classe.mkdir(exist_ok=True)

            # Sposta il file nella cartella
            shutil.move(str(file), str(cartella_classe / file.name))
            print(f"Spostato '{file.name}' in '{cartella_classe}'")

organizza_file_per_classe(percorso)
```

```python
# Percorso di D2
percorso_d2 = "Dataset/D2tofix/dataset"

classi_d2 = estrai_classi_cartelle(percorso_d2)
print(len(classi_d2))
print(classi_d2)
```

Controllo le classi in comune tra D1 e D2

```python
classi_d1d2 = list(set(classi_d1) & set(classi_d2))
print(f"Classi in comune tra D1 e D2: {len(classi_d1d2)}")
print(classi_d1d2)

classi_d2md1 = list(set(classi_d2) - set(classi_d1))
print(f"Classi presenti in D2 ma non in D1: {len(classi_d2md1)} ")
print(classi_d2md1)
```

Ci sono delle classi in D2 che non sono in D1.
Questo pu essere un problema dato che D1 ha molte pi istanze di D2.

Prendo ogni classe di D3 e rinomino le cartelle con le etichette relative.

```python
mapping_file = Path("Dataset/piecesMapping.CSV").resolve()
percorso_cartelle = Path("Dataset/Cropped Images").resolve()

name_to_code = {}
with open(mapping_file, "r") as f:
    reader = csv.reader(f)
    next(reader)  # Salta l'intestazione
    for row in reader:
        name_to_code[row[0].lower()] = row[1]

# Rinominare le cartelle nel percorso specificato
def rinomina_cartelle(percorso, mapping):
    for nome_cartella in os.listdir(percorso):
        percorso_completo = os.path.join(percorso, nome_cartella)

        # Controlla che sia una cartella
        if os.path.isdir(percorso_completo):
            nome_cartella_lower = nome_cartella.lower()

            # Trova il codice corrispondente al nome
            if nome_cartella_lower in mapping:
                nuovo_nome = mapping[nome_cartella_lower]
                nuovo_percorso = os.path.join(percorso, nuovo_nome)

                # Rinomina la cartella
                os.rename(percorso_completo, nuovo_percorso)
                print(f"Cartella '{nome_cartella}' rinominata in '{nuovo_nome}'")
            else:
                print(f"Nessun codice trovato per '{nome_cartella}'")

# Esempio di utilizzo

rinomina_cartelle(percorso_cartelle, name_to_code)
```

Le classi in D3 sono tutte presenti in D1 tranne '6141'.

Se decidessi di utilizzare D3 dovrei trovare un modo per bilanciare le classi.

Controllo a questo punto le classi in comune nei 3 dataset e quelle esclusive di ciascun dataset.

```python
# Classi in comune tra tutti
classi_d1d2d3 = list(set(classi_d1) & set(classi_d2) & set(classi_d3))
print(f"Classi in comune tra D1, D2 e D3: {len(classi_d1d2d3)}")
print(classi_d1d2d3)
# Classi esclusive di d1
classi_d1excl = list(set(classi_d1) - set(classi_d2) - set(classi_d3))
print(f"Classi esclusive di D1: {len(classi_d1excl)}")
print(classi_d1excl)
# Classi esclusive di d2
classi_d2excl = list(set(classi_d2) - set(classi_d1) - set(classi_d3))
print(f"Classi esclusive di D1: {len(classi_d2excl)}")
print(classi_d2excl)
# Classi esclusive di d3
classi_d3excl = list(set(classi_d3) - set(classi_d1) - set(classi_d2))
print(f"Classi esclusive di D1: {len(classi_d3excl)}")
print(classi_d3excl)
```
